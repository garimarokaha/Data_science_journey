```python
#library management system
# 3 classes => Book , member, library


#defining class Book
class Book:
    def __init__(self, title , author, Book_no):
        self.title = title
        self.author = author
        self.Book_no = Book_no
        self.is_borrowed = False # book lageko cha ta pailei bahneraa sodhha false
        #book ko availability check garnaaa

    def __str__(self):
        return f"title:{ self.title} , author : {self.author}, Book_no : {self.Book_no}"
        

#defining class member
class Member:
    def __init__(self,member_id,name):
        self.member_id = member_id
        self.name = name
        self.borrowed_book =[] #list to store the borrowed book by the member (member le lageko book dekhauna lai)


    def borrow_book(self,book):#book borrow garera laijana
        if book.is_borrowed:
            print(f"The book '{book.title}' is already borrowed ")
        else:
            self.borrowed_book.append(book)
            book.is_borrowed = True
            print(f"{self.name} borrowed '{book.title}'.")

    def return_book(self,book): #lageko book firta garna ko lagi
       if book in self.borrowed_book:
           self.borrowed_book.remove(book)
           book.is_borrowed = False 
           print(f"{self.name} returned '{book.title}'.")
       else:
           print(f"{self.name } has not returned '{book.title}'.")

           
           
#defining class library
class Library:
    def __init__(self):
        self.books =[] # list  to store all kind of books in the library
        self.members = [] # list to store all kind of members of the library

    #library ma book thapna ko lagi
    def add_book(self,book):
        self.books.append(book)
        print(f"{book.title} added in the library")
        
    # library ma member thapna (add garna )ko lagi
    def add_member(self,member):
        self.members.append(member)
        print(f"{member.name} added in the library")


    #vako book dekhaunaa
    def show_available_books(self):
        available_books = [book for book in self.books if not book.is_borrowed]
        
        if available_books:
            print("\nAvailable Books:")
            for book in available_books:
                print(book)
        else:
            print("No books available at the moment.")
        
 #main program

#create books
book1 = Book("saya","subin bhattrai",1)
book2 = Book("karnali blues ","buddhisagar",2)
book3 = Book("yek sarko maya","gs pousel", 3)

#create  members
member1= Member(1,"Jolly")
member2= Member(2,"dolly")

#create library and add books and library
library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_member(member1)
library.add_member(member2)

# Member borrows books
member1.borrow_book(book1)
member2.borrow_book(book2)

# Show available books again after borrowing
library.show_available_books()

# Member returns a book
member1.return_book(book1)

# Show available books after returning
library.show_available_books()
```
